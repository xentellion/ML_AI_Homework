```python
increase = df['2020_census'] - df['2010_census']
df.insert(loc = 5, column='pop_increase', value=increase)
df.sort_values(by='pop_increase', ascending = True).head()
# Ага! Они указывают прирост в абсолютных величинах. Детройту, конечно, от этого лучше, но все же
```
